```python
#Filter to keep only movies in the 90s
df_90s = df[(df['release_year'] >= 1990) & (df['release_year'] < 2000)]
df_90s.head()

```

```python
# Visualize the duration column based on df_90s

plt.hist(df_90s['duration'])
#Label 
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Number of Movies')

plt.show()
```

```python
#Filter the data again to keep only the action movies

df_90s['genre'].unique()
```

```python
df_90s_action =  df_90s[df_90s['genre'] == 'Action']
df_90s_action.head()
```

```python
#Use a for loop and a counter to count how many short action movies there are in the 1990s

# Lets assume our short movies are less than 90 mins

shortmv_count = 0

for label, row in df_90s_action.iterrows():
    if row["duration"] < 90:
        shortmv_count = shortmv_count + 1
    else:
        shortmv_count = shortmv_count
print(shortmv_count)
```
